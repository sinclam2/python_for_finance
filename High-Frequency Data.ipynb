http://hopey.netfonds.no/paperhistory.php?paper=STL.OSE&csv_format=csv

```python
df_STL = pd.read_csv('STL_hf_data.csv')
```

```python
df_STL['Year']=df_STL['time'].map(lambda x: x[:4])
df_STL['Month']=df_STL['time'].map(lambda x: x[4:6])
df_STL['Day']=df_STL['time'].map(lambda x: x[6:8])
df_STL['Hour']=df_STL['time'].map(lambda x: x[9:11])
df_STL['Minute']=df_STL['time'].map(lambda x: x[11:13])
df_STL['Second']=df_STL['time'].map(lambda x: x[13:15])
df_STL['plot-axis'] = (df_STL['Hour'].astype(str)+df_STL['Minute'].astype(str)+df_STL['Second'].astype(str))
```

```python
df_STL.plot(x='plot-axis',y='bid', figsize=(30,15))
```
